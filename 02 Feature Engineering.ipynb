## <b> Carga de librerias </b>

```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
import modules
import preprocessors as pp 
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler 
```

## <b> Carga de dataset </b>

```python
datos = pd.read_csv(r'dataset procesados/feature.txt', sep=';')
datos.drop(columns=['Good', 'Poor', 'Standard'], inplace=True)
datos.head()
```

## <b> Aplicacion de Pipeline de Feature Engineering </b>

```python
def instanciaPipeline(dataset, var_y):
    
    categoric_vars, discrete_vars, continues_vars = modules.getColumnsDataTypes(df=dataset)
    categoric_vars.remove(var_y)


    feature_engineering_pipeline = Pipeline(steps=[
        ('Categorical_Encoder', 
            pp.CategoricalEncoderOperator(varNames=categoric_vars)
        )
    ])

    return feature_engineering_pipeline

procesed_data = instanciaPipeline(dataset=datos, var_y='Credit_Score').fit_transform(datos)
```

```python
procesed_data.to_csv('dataset procesados/modelado.txt', sep=';', index=False)
```
